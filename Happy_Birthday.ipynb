```python
import matplotlib.pyplot as plt

# Number of possible birthdays
days = 365

def birthday_probability(n, days=365):
    """
    Computes the probability that at least two people
    share the same birthday among n people.
    """
    prob_all_different = 1.0
    for k in range(n):
        prob_all_different *= (days - k) / days
    return 1 - prob_all_different

n_chosen = 23
max_people = 100


# Compute probabilities for plotting
people = list(range(1, max_people + 1))
probabilities = [birthday_probability(n) for n in people]

# Probability at chosen n
p_chosen = birthday_probability(n_chosen)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(people, probabilities, linewidth=4)

plt.axvline(n_chosen, linestyle='--', linewidth=3)
plt.axhline(p_chosen, linestyle='--', linewidth=3)

plt.xlabel("Number of people", fontsize=30)
plt.ylabel("$P$(shared birthday)", fontsize=30)

plt.xticks(fontsize=30)
plt.yticks(fontsize=30)

plt.tight_layout()
plt.savefig("birthday.pdf")
plt.show()

# Print numerical result
print(f"For n = {n_chosen}, probability = {p_chosen:.4f}")
```

```python
import matplotlib.pyplot as plt

days = 365

def shared_birthday_probability(n):
    """Probability that at least two people share a birthday"""
    prob_all_different = 1.0
    for k in range(n):
        prob_all_different *= (days - k) / days
    return 1 - prob_all_different


def match_specific_person_probability(n):
    """Probability that someone matches a specific person's birthday"""
    return 1 - ((days - 1) / days) ** n

max_people = 400

# Compute values
people = list(range(1, max_people + 1))
shared_probs = [shared_birthday_probability(n) for n in people]
specific_probs = [match_specific_person_probability(n) for n in people]

# Plot
plt.figure(figsize=(9, 6))
plt.plot(
    people,
    shared_probs,
    linewidth=4,
    label="Any shared birthday",
)
plt.plot(
    people,
    specific_probs,
    linestyle="--",
    linewidth=4,
    label="Match with fixed person"
)

plt.xlabel("Number of people", fontsize=(30))
plt.ylabel("Probability", fontsize=(30))

plt.xticks(fontsize=25)
plt.yticks(fontsize=25)

plt.legend(fontsize=25)

plt.tight_layout()
plt.savefig("birthday2.pdf")
plt.show()
```
